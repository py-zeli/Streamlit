```python
import requests
import time
import random

# URL do seu endpoint Flask para receber coordenadas
URL_ENDPOINT = "http://127.0.0.1:5000/nova-coordenada"

# Lista de coordenadas aproximadas de grandes cidades do Brasil
CIDADES = {
    "São Paulo": (-23.5505, -46.6333),
    "Rio de Janeiro": (-22.9068, -43.1729),
    "Brasília": (-15.7797, -47.9297),
    "Salvador": (-12.9777, -38.5016),
    "Fortaleza": (-3.7319, -38.5267),
    "Belo Horizonte": (-19.9167, -43.9345),
    "Manaus": (-3.1190, -60.0217),
    "Curitiba": (-25.4284, -49.2733),
    "Recife": (-8.0578, -34.8824),
    "Porto Alegre": (-30.0346, -51.2177),
    "Belém": (-1.4558, -48.5035)
}

# Loop infinito para enviar requisições
while True:
    try:
        # Escolhe uma cidade aleatoriamente da lista
        nome_cidade, coords_cidade = random.choice(list(CIDADES.items()))

        # Gera uma coordenada ligeiramente aleatória próxima à cidade
        latitude_aleatoria = coords_cidade[0] + random.uniform(-0.5, 0.5)
        longitude_aleatoria = coords_cidade[1] + random.uniform(-0.5, 0.5)

        # Dados a serem enviados no corpo da requisição
        dados = {
            "latitude": latitude_aleatoria,
            "longitude": longitude_aleatoria
        }

        # Envia a requisição POST para o servidor Flask
        response = requests.post(URL_ENDPOINT, json=dados)

        # Verifica se a requisição foi bem-sucedida
        if response.status_code == 200:
            print(f"Requisição enviada com sucesso para: {nome_cidade} ({latitude_aleatoria:.4f}, {longitude_aleatoria:.4f})")
        else:
            print(f"Erro ao enviar requisição. Código de status: {response.status_code}")
            print(f"Resposta do servidor: {response.text}")

    except requests.exceptions.RequestException as e:
        print(f"Erro de conexão: {e}")
        
    # Espera 1 segundo antes de enviar a próxima requisição
    time.sleep(0.01)
```
